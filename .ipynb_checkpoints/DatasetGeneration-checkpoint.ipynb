```python
import os, glob, re
import pandas as pd

base = "data/inference_outputs"
for b in os.listdir(base):
    d = os.path.join(base, b)
    if not os.path.isdir(d): continue
    files = glob.glob(os.path.join(d, "*.csv"))
    files.sort(key=lambda p: int(re.search(r"batch_(\d+)", p).group(1)) if re.search(r"batch_(\d+)", p) else float("inf"))
    df = pd.concat((pd.read_csv(p) for p in files), ignore_index=True)
    df.to_csv(os.path.join(d, "combined.csv"), index=False)
```

```python
import os
import pandas as pd
import json

base_dir = "data/inference_outputs"

for benchmark in os.listdir(base_dir):
    bench_dir = os.path.join(base_dir, benchmark)
    combined_path = os.path.join(bench_dir, "combined.csv")

    if not os.path.isdir(bench_dir) or not os.path.exists(combined_path):
        continue

    df = pd.read_csv(combined_path)

    # Detect whether to use normalized or raw accuracy
    if all(col in df.columns for col in ["acc_norm_small", "acc_norm_medium", "acc_norm_large"]):
        acc_cols = {
            "small": "acc_norm_small",
            "medium": "acc_norm_medium",
            "large": "acc_norm_large"
        }
    elif all(col in df.columns for col in ["acc_small", "acc_medium", "acc_large"]):
        acc_cols = {
            "small": "acc_small",
            "medium": "acc_medium",
            "large": "acc_large"
        }
    else:
        print(f"Skipping {benchmark} — no suitable accuracy columns found.")
        continue

    if "input_text" not in df.columns:
        print(f"Skipping {benchmark} — missing 'input_text' column.")
        continue

    sample_size = min(100, len(df))
    sampled = df.sample(n=sample_size, random_state=42)

    json_list = []
    for _, row in sampled.iterrows():
        entry = {
            "question": f"Question:{row['input_text']}\nAnswer:",
            "scores": {
                size: float(row[acc_cols[size]]) for size in ["small", "medium", "large"]
            },
            "cluster_id": 0
        }
        json_list.append(entry)

    output_path = os.path.join(bench_dir, f"{benchmark}_train.json")
    with open(output_path, "w") as f:
        json.dump(json_list, f, indent=2)

    print(f"Saved: {output_path}")
```
